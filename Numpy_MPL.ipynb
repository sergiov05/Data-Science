```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5,6],dtype=np.int32)
```

```python
y = np.pow(x,2)
```

```python
z = np.sqrt(x)
```

```python
fig, axs = plt.subplots(1,2)
#setting first graph
axs[0].plot(x,y)
axs[0].grid(True)
axs[0].set_title("X2")

#Setting second graph
axs[1].plot(x,z)
axs[1].set_title("sqrt(X)")
axs[1].grid(True)

plt.show()
```
